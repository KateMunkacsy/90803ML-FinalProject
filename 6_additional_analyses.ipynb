```python
import pandas as pd
```

```python
demo_df = pd.read_excel("Data/ACS 5 Year_Demographic Data_2019-2023.xlsx", sheet_name = "Data", header = [0, 1])
demo_df.head()
```

```python
demo_df.columns = demo_df.columns.map('_'.join)
demo_df.rename({'Unnamed: 0_level_0_Label':'Label'}, inplace = True, axis = 1)
demo_df.head()
```

```python
import numpy as np
demo_df['Topic'] = np.where(demo_df.iloc[:, 1:].isnull().all(axis=1), demo_df['Label'], np.nan)
demo_df.insert(0, 'Topic', demo_df.pop('Topic'))
demo_df['Topic'] = demo_df['Topic'].ffill()
demo_df.head()
```

```python
demo_df = demo_df[~demo_df.iloc[:, 2:].isnull().all(axis=1)].copy()
demo_df
```

```python
demo_df.drop_duplicates(inplace = True)
demo_df = demo_df.groupby(["Topic", "Label"]).first()
demo_df = demo_df.reset_index()
```

```python
piv = pd.wide_to_long(demo_df, stubnames=["Estimate", "Margin of Error", "Percent"], i = ["Topic", "Label"],  j="Location", sep="_").reset_index()
#demo_df = pd.wide_to_long(demo_df, i=['Unnamed: 0_level_0_Label'], j='location', suffix = )
piv.head()
```
