```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Convert date column to datetime
ridership['month_start'] = pd.to_datetime(ridership['month_start'])

# Plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=ridership, x='month_start', y='avg_riders', hue='route', errorbar=None)
plt.legend().remove()

plt.title("Bus Ridership Over Time by Line")
plt.xlabel("Date")
plt.ylabel("Ridership")
plt.tight_layout()
plt.show()
```

```python
ridership_filtered = ridership[ridership['month_start'] > '2023-01-01']
```

```python
# Plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=ridership_filtered, x='month_start', y='avg_riders', hue='route', errorbar=None)
plt.legend().remove()

plt.title("Bus Ridership Over Time by Line")
plt.xlabel("Date")
plt.ylabel("Ridership")
plt.tight_layout()
plt.show()
```
