```python
data = pd.read_csv('avg_ridership.csv')
```

```python
data.head(20)
```

```python
data.drop(['_id', 'ridership_route_code', 'route_full_name', 'mode', 'month_start', 'day_count'], axis=1, inplace=True)
```

```python
data
```

```python
data.dtypes
```

```python
data['year_month'] = data['year_month'].astype(str)
```

```python
data['year'] = data['year_month'].str[:4]
data['month'] = data['year_month'].str[4:]
```

```python
data['year'] = data['year'].astype(int)
data['month'] = data['month'].astype(int)
data.drop('year_month', axis=1, inplace=True)
```

```python
data
```

```python
# Step 1: Create a datetime column
data['date'] = pd.to_datetime(data[['year', 'month']].assign(day=1))

# Step 2: Sort by bus line and date
data = data.sort_values(['route', 'date'])

# Step 3: Create the lagged column
data['lagged_avg_riders'] = data.groupby('route')['avg_riders'].shift(1)
```

```python
data
```

```python
data.isna().sum()
```

```python
data.dropna(subset=['route'], inplace=True)
data
```

```python
# Assuming month is an integer column (1 = Jan, 12 = Dec)
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Fall'

# Apply to your dataframe
data['season'] = data['month'].apply(get_season)
```
