```python
data = pd.read_csv('stop_usage.csv')
```

```python
data.head().T
```

```python
data.drop(['clever_id', 'stop_name', 'mode', 'stop_type', 'days'], axis=1, inplace=True)
```

```python
data.head()
```

```python
data['datekey'] = data['datekey'].astype(str)
```

```python
data['year'] = data['datekey'].str[:4]
data['month'] = data['datekey'].str[4:]
```

```python
data.head()
```

```python
data['year'] = data['year'].astype(int)
data['month'] = data['month'].astype(int)
data.drop('datekey', axis=1, inplace=True)
```

```python
# Step 1: Create a datetime column
data['date'] = pd.to_datetime(data[['year', 'month']].assign(day=1))

# Step 2: Sort by route, service_day, and date
data = data.sort_values(['stop_id', 'route_name', 'serviceday', 'date'])

# Step 3: Create the lagged column (lagging within route + service_day group)
data['lagged_avg_ons'] = data.groupby(['stop_id', 'route_name', 'serviceday'])['avg_ons'].shift(1)
```
